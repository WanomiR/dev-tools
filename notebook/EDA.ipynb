```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('vehicles_us.csv')
```

```python
df.head()
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
df['model_year'] = df['model_year'].astype('Int64')
df['cylinders'] = df['cylinders'].astype('Int64')
```

```python
df['is_4wd'] = df['is_4wd'].notnull()
```

```python
# Histogram for price
df['price'].hist(bins=100)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Prices')
plt.xlim(0, 100000)
plt.show()

# Bar chart for fuel type
df['fuel'].value_counts().plot(kind='bar')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.title('Frequency of Fuel Types')
plt.show()
```

```python
# Scatter plot for price vs. odometer
plt.scatter(df['odometer'], df['price'])
plt.xlabel('Odometer')
plt.ylabel('Price')
plt.title('Price vs. Odometer')
plt.show()

# Box plot for cylinders vs. price
sns.boxplot(x='cylinders', y='price', data=df)
plt.xlabel('Cylinders')
plt.ylabel('Price')
plt.title('Price Distribution by Cylinder Count')
plt.show()
```
